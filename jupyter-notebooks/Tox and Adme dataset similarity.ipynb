```python

#df = pd.read_csv('/home/gareth/src/os-models/data/vault/adme_tox_cmc.csv.gz')
df = pd.read_csv('/home/gareth/src/os-models/data/vault/adme_tox_nci.csv.gz')
```

```python
def rename_columns(df):
    def rename_col(no, n):
        new_n = re.sub(r" \[.*\]$", "", n)
        return new_n
    columns_map = {n:rename_col(i, n) for i, n in enumerate(df.columns)}
    df.rename(columns=columns_map, inplace=True)
    return df

df = rename_columns(df)
df = df.drop(['ID'], axis=1)
```
