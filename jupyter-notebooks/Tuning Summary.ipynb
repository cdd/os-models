```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as n

import sys
sys.path.insert(0, 'python')
from tuning_summary import TuningSummary
```

```python
classification=False
all_systems=True
#subset=None
#subset='spark'
subset='regular'

tuning_summary = TuningSummary(classification=classification, all_systems=all_systems, subset=subset)

metric = tuning_summary.metric_name()
global_col = '{}_global'.format(metric)
common_col = '{}_common'.format(metric)

summary_df=tuning_summary.summary_df
if classification and not all_systems:
    summary_df = summary_df[~summary_df['name'].isin(['data_bubonic', 'data_kcnq'])]
summary_df = summary_df.sort_values(['name', 'method'])
summary_df['method'] = summary_df['method'].astype("category")
summary_df
```

```python
rmse_min_summary = summary_df.groupby(['method'])[global_col].mean().sort_values()
rmse_min_summary
```

```python
metric_common_summary = summary_df.groupby(['method'])[common_col].mean().sort_values()
metric_common_summary
```

```python
print(metric_common_summary.index.tolist())
summary_df['method'].cat.set_categories(metric_common_summary.index.tolist(), inplace=True)
summary_df.sort_values(['method'])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20, 12))
#sns.set(font_scale=1.2)
if classification:
    axes[0].set(ylim=(0.5,1.0))
    axes[1].set(ylim=(0.5,1.0))
else:
    axes[0].set(ylim=(0.4,1.0))
    axes[1].set(ylim=(0.2,1.3))  
axes[0].get_xaxis().set_visible(False) 
axes[0].tick_params(axis='y', labelsize=15)
sns.boxplot(x='method', y=common_col, data=summary_df, ax=axes[0])
for tick in axes[1].get_xticklabels():
    tick.set_rotation(80)
axes[1].tick_params(axis='both', labelsize=15)
sns.pointplot(x='method', y=common_col, hue='name', data=summary_df, ax=axes[1])
```

```python

height = 8 if all_systems else 16
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, height))
if classification:
    axes[0].set(xlim=(0.6,1.0))
    axes[1].set(xlim=(0.5,1.0))
else:
    axes[0].set(xlim=(0.4,1.0))
    axes[1].set(xlim=(0.2,1.3))  
axes[0].tick_params(axis='both', labelsize=15)
sns.boxplot(y='method', x=common_col, data=summary_df, orient='h', ax=axes[0])
axes[1].get_yaxis().set_visible(False)
axes[1].tick_params(axis='x', labelsize=15)
sns.pointplot(y='method', x=common_col, hue='name', data=summary_df, orient='h', ax=axes[1])
plt.setp(axes[1].get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(axes[1].get_legend().get_title(), fontsize='15') 
```
